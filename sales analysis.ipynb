## Sales Data Analysis Using Python

Introduction
This project focuses on analyzing sales data using Python to uncover trends, patterns, and business insights.
The analysis includes data cleaning, aggregation, exploratory data analysis, and visualization to support
data-driven decision-making.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Loading the Dataset
The sales dataset is loaded using Pandas for further analysis.

```python
df = pd.read_excel(r"C:\Users\user\Documents\portfolio docs\Sales_Performance_Dataset.xlsx")
df.head()
```

```python
df.info()
```

## Data Cleaning
This step ensures data accuracy by handling missing values, correcting data types,
and preparing the dataset for analysis.


```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop missing values (or fill if needed)
df = df.dropna()
```

```python
# Convert date column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## Exploratory Data Analysis (EDA)
EDA helps in understanding overall sales performance, distribution,
and key contributors such as regions and products.

```python
df.describe()
```

```python
# Total sales
df['Sales Amount'].sum()
```

```python
# Sales by region
sales_by_region = df.groupby('Region')['Sales Amount'].sum().sort_values(ascending=False)
sales_by_region
```

```python
# Sales by product
sales_by_product = df.groupby('Product')['Sales Amount'].sum().sort_values(ascending=False)
sales_by_product
```

## Data Visualization
Visualizations are used to clearly communicate trends and performance insights.

```python
sales_by_region.plot(kind='bar')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
sales_over_time = df.groupby('Order Date')['Sales Amount'].sum()

sales_over_time.plot(figsize=(10,5))
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

## Insights & Conclusion

### Key Insights
- Certain regions consistently generate higher sales than others.
- Sales trends show seasonal patterns over time.
- A small number of products contribute a large portion of total revenue.

### Conclusion
Python proved effective for cleaning, analyzing, and visualizing sales data.
This analysis can help businesses identify high-performing regions and products,
optimize strategies, and improve decision-making.
